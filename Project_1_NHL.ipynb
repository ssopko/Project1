```python
#concatinate date
raw_dataframe = [Ducks_df, Coyotes_df, Bruins_df, Sabres_df, Flames_df, Hurricanes_df, 
                 Blackhawks_df, Avalance_df, Blue_Jackets_df, Stars_df, Red_Wings_df, Oilers_df,
                 Panthers_df, Kings_df, Wild_df, Canadiens_df, Predators_df, Devils_df, 
                 Islanders_df, Rangers_df, Senators_df, Flyers_df, Penguins_df, Sharks_df,
                 Blues_df, Lighting_df, Maple_leaves_df, Canucks_df,Golden_Knights_df,
                 Capitals_df,Jets_df]

NHL_dataframe = pd.concat(raw_dataframe)
NHL_dataframe.head()
```

```python
NHL_dataframe["Winning %"] = NHL_dataframe["W"] / NHL_dataframe["GP"]
NHL_dataframe.head()
```

```python
#rename columns
NHL_dataframe = NHL_dataframe.rename(columns={"Lg": "League", "GP": "Games Played", "W": "Wins",
                                              "L": "Losses", "T": "Ties", "OL": "Overtime Losses"
                                             })
NHL_dataframe.head()
```

```python
#Nix irrelevant columns and remove asterisks
NHL_dataframe = NHL_dataframe[["Season", "League", "Team", "Games Played", "Wins", "Losses", "Ties", "Overtime Losses", "Winning %"]]
NHL_dataframe["Team"] = NHL_dataframe["Team"].str.replace('*', '')
NHL_dataframe.head()
```

```python
#pivottime
pivot = NHL_dataframe.pivot(values = "Winning %", index = "Team", columns = "Season")
pivot.head()
```

```python
pivot_2000_01 = pivot.sort_values(["2000-01"], ascending = False)
pivot_2000_01 = pivot_2000_01["2000-01"].head()

pivot_2001_02 = pivot.sort_values(["2001-02"], ascending = False)
pivot_2001_02 = pivot_2001_02["2001-02"].head()

pivot_2002_03 = pivot.sort_values(["2002-03"], ascending = False)
pivot_2002_03 = pivot_2002_03["2002-03"].head()

pivot_2003_04 = pivot.sort_values(["2003-04"], ascending = False)
pivot_2003_04 = pivot_2003_04["2003-04"].head()

pivot_2005_06 = pivot.sort_values(["2005-06"], ascending = False)
pivot_2005_06 = pivot_2005_06["2005-06"].head()

pivot_2006_07 = pivot.sort_values(["2006-07"], ascending = False)
pivot_2006_07 = pivot_2006_07["2006-07"].head()

pivot_2007_08 = pivot.sort_values(["2007-08"], ascending = False)
pivot_2007_08 = pivot_2007_08["2007-08"].head()

pivot_2008_09 = pivot.sort_values(["2008-09"], ascending = False)
pivot_2008_09 = pivot_2008_09["2008-09"].head()

pivot_2009_10 = pivot.sort_values(["2009-10"], ascending = False)
pivot_2009_10 = pivot_2009_10["2009-10"].head()

pivot_2010_11 = pivot.sort_values(["2010-11"], ascending = False)
pivot_2010_11 = pivot_2010_11["2010-11"].head()

pivot_2011_12 = pivot.sort_values(["2011-12"], ascending = False)
pivot_2011_12 = pivot_2011_12["2011-12"].head()

pivot_2012_13 = pivot.sort_values(["2012-13"], ascending = False)
pivot_2012_13 = pivot_2012_13["2012-13"].head()

pivot_2013_14 = pivot.sort_values(["2013-14"], ascending = False)
pivot_2013_14 = pivot_2013_14["2013-14"].head()

pivot_2014_15 = pivot.sort_values(["2014-15"], ascending = False)
pivot_2014_15 = pivot_2014_15["2014-15"].head()

pivot_2015_16 = pivot.sort_values(["2015-16"], ascending = False)
pivot_2015_16 = pivot_2015_16["2015-16"].head()

pivot_2016_17 = pivot.sort_values(["2016-17"], ascending = False)
pivot_2016_17 = pivot_2016_17["2016-17"].head()

pivot_2017_18 = pivot.sort_values(["2017-18"], ascending = False)
pivot_2017_18 = pivot_2017_18["2017-18"].head()

top_winning = []
top_winning = pivot_2000_01.reset_index()
seasons = [pivot_2001_02, pivot_2002_03, pivot_2003_04, pivot_2005_06,
           pivot_2006_07, pivot_2007_08, pivot_2008_09, pivot_2009_10, pivot_2010_11,
           pivot_2011_12, pivot_2012_13, pivot_2013_14, pivot_2014_15, pivot_2015_16,
           pivot_2016_17, pivot_2017_18]
```

```python
for top in seasons:
    top_winning = pd.concat([top_winning, top.reset_index()], axis = 1)

top_winning = top_winning[["2000-01", "2001-02", "2002-03", "2003-04", "2005-06", "2006-07",
                           "2007-08", "2008-09", "2009-10", "2010-11", "2011-12", "2012-13",
                           "2013-14", "2014-15", "2015-16", "2016-17", "2017-18"]]
top_winning_csv = top_winning.to_csv("Top 5 Winning Percentages By Year.csv", index = False, header = True)
top_winning
```
