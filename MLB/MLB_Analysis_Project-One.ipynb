```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
teams = ["Angels", "Astros", "Athletics", "Blue Jays", "Braves", "Brewers", "Cardinals", "Cubs", "Diamondbacks", "Dodgers", "Giants", "Indians","Mariners", "Marlins", "Mets", "Nationals","Orioles", "Padres", "Phillies", "Pirates", "Rangers", "Rays", "Red_Sox", "Reds", "Rockies", "Royals", "Tigers", "Twins", "White_Sox", "Yankees"]
years = ["Index", 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
index_holders = ["a","b","c",'d', "e", "f", "g","h","i", "j", "k",'l','m','n', 'o', 'p', 'q', 'r', 's', 't', 'u','v', 'w', 'x', 'y', 'z', 'aa', 'ab', 'ac','ad']
```

```python
df={}
for i in range(len(teams)):
    df[teams[i]] = pd.read_csv(f'{teams[i]}.csv', index_col = "Year")
df[teams[0]].head()
```

```python
new_df = pd.DataFrame(columns=years)
new_df["Index"]= index_holders
new_df.head(2)
```

```python
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
list = []
list_all = []
for year in years:
    for i in range(len(teams)):
        list.append(df[teams[i]]["W-L%"][year])
        list_all.append(df[teams[i]]["W-L%"][year])
    list.sort(reverse=False)
    new_df[year] = list
    list = []
new_df.head(10)
```

```python
new_df = new_df.drop(["Index"], axis =1)
new_df.head()
```

```python
new_df.describe()
```

```python
worst_df = new_df.head(5)
worst_df
```

```python
worst_df.describe()
```

```python
worst_df.to_csv("worst_baseball.csv", index=False, encoding='utf-8')
```

```python
worst_df = worst_df.T
worst_df
```

```python
worst_graph = plt.plot(worst_df, marker = "o", alpha = 325)
labels = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008,
                         2009, 2010,2011, 2012, 2013, 2014, 2015, 2016, 2017]
plt.grid()
plt.xticks(labels, rotation = 45)
plt.xlabel("Year")
plt.ylabel("Winning %")
plt.title("MLB Bottom 5 Winning % By Year")
plt.savefig("MLB Bottom 5 Winning % By Year")
plt.show()
```
