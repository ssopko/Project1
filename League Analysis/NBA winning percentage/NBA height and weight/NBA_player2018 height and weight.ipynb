```python
import csv
import matplotlib.pyplot as plt
import openweathermapy as ow
import pandas as pd
import requests
import json
import numpy as np
from pandas import DataFrame
```

```python
url = "http://data.nba.net/10s/prod/v1/2018/players.json"
response  = requests.get(url)
data = response.json()
print(data)
```

```python
data_playerinfo = data['league']['standard']
data_playerinfo
```

```python
df = DataFrame(data_playerinfo)
df
```

```python
df.dropna(subset=["heightMeters"])
height_group = df.groupby("heightMeters")
count_height_group  = height_group["heightMeters"].count()
count_height_group
count_chart1 = count_height_group.plot(kind='bar')

# Save the figure to the output folder
#plt.savefig("output_data/latVwind.png")
```

```python
df.dropna(subset=["country"])
country_group = df.groupby("country")
count_country_group  = country_group["country"].count()
count_country_group
count_chart = count_country_group.plot(kind='bar')
```

```python
df.dropna(subset=["pos"])
height_group = df.groupby("pos")
count_height_group  = height_group["heightMeters"].mean()
count_height_group
```
