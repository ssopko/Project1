```python
import base64
import requests


def send_request():
    # Request

    try:
        response = requests.get(
            url="https://api.mysportsfeeds.com/v2.0/pull/nba/injuries.json",
            params={
                "fordate": "20161121"
            },
            headers={
                "Authorization": "Basic " + api_key},'MYSPORTSFEEDS').encode('utf-8')).decode('ascii')
                #"Authorization": "Basic " + base64.b64encode('{}:{}'.format({apikey_token},MYSPORTSFEEDS).encode('utf-8')).decode('ascii')
            }
        )
        print("Authorization: Basic " + base64.b64encode('{}:{}'.format({api_key},'MYSPORTSFEEDS').encode('utf-8')).decode('ascii'))
        #print(requests.headers)
        #print('Authorization:Basic ' + api_key + ":" + "MYSPORTSFEEDS")
        print('Response HTTP Status Code: {status_code}'.format(
            status_code=response.status_code))
        print('Response HTTP Response Body: {content}'.format(
            content=response.content))
    except requests.exceptions.RequestException:
        print('HTTP Request failed')
```

```python
send_request()
```
