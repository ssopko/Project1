```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Get all NFL Excel Files into a DF
az_cardinals = pd.read_excel('AZ Cardinals.xlsx')

atl_falcons = pd.read_excel('ATL Falcons.xlsx')  

bal_ravens = pd.read_excel('BAL Ravens.xlsx')

buf_bills = pd.read_excel('BUF Bills.xlsx')

car_panthers = pd.read_excel('CAR Panthers.xlsx')

chi_bears = pd.read_excel('CHI Bears.xlsx')

cin_bangels = pd.read_excel('CIN Bangels.xlsx')

cle_browns = pd.read_excel('CLE Browns.xlsx')

dal_cowboys = pd.read_excel('DAL Cowboys.xlsx')

den_broncos = pd.read_excel('DEN Broncos.xlsx')

det_lions = pd.read_excel('DET Lions.xlsx')

gb_packers = pd.read_excel('GB Packers.xlsx')

hou_texans = pd.read_excel('HOU Texans.xlsx')

ind_colts = pd.read_excel('IND Colts.xlsx')

jac_jaguars = pd.read_excel('JAC Jaguars.xlsx')

kc_chiefs = pd.read_excel('KC Chiefs.xlsx')

la_chargers = pd.read_excel('LA Chargers.xlsx')

la_rams = pd.read_excel('LA Rams.xlsx')

mia_dolphins = pd.read_excel('MIA Dolphins.xlsx')

min_vikings = pd.read_excel('MIN Vikings.xlsx')

ne_patriots = pd.read_excel('NE Patriots.xlsx')

no_saints = pd.read_excel('NO Saints.xlsx')

ny_giants = pd.read_excel('NY Giants.xlsx')

ny_jets = pd.read_excel('NY Jets.xlsx')

oak_raiders = pd.read_excel('OAK Raiders.xlsx')

phi_eagles = pd.read_excel('PHI Eagles.xlsx')

pitt_steelers = pd.read_excel('PITT Steelers.xlsx')

sea_seahawks = pd.read_excel('SEA Seahawks.xlsx')

sf_49ers = pd.read_excel('SF 49ers.xlsx')

tb_buccaneers = pd.read_excel('TB Buccaneers.xlsx')

ten_titans = pd.read_excel('TEN Titans.xlsx')

was_redskins = pd.read_excel('WAS Redskins.xlsx')
```

```python
#Create a DF to store all the NFL Data starting with the first team
nfl_data = az_cardinals
#Create a list with the res of the team DFs to iterate through for appending
nfl_teams = [atl_falcons, bal_ravens,buf_bills, car_panthers, chi_bears, cin_bangels, cle_browns,
            dal_cowboys, den_broncos, det_lions, gb_packers, hou_texans, ind_colts, jac_jaguars,
            kc_chiefs, la_chargers, la_rams, mia_dolphins, min_vikings, ne_patriots, no_saints,
            ny_giants, ny_jets, oak_raiders, phi_eagles, pitt_steelers, sf_49ers, sea_seahawks,
            tb_buccaneers, ten_titans, was_redskins]
```

```python
#loop through the list to create a massive DF with all data
for teams in nfl_teams:
    nfl_data = pd.concat([nfl_data, teams], axis = 0)
nfl_data_csv = nfl_data.to_csv("Total NFL Data.csv", index = False, header = True)
nfl_data.head()
```

```python
#Create winning % column
nfl_data["Winning %"] = nfl_data["W"] / (nfl_data["W"] + nfl_data["L"] + nfl_data["T"])
nfl_data.head()
```

```python
#Rename the columns I need
nfl_data = nfl_data.rename(columns={"Lg": "League", "Tm": "Team", "W": "Wins", "L": "Loses", "T": "Ties"})
nfl_data.head()
```

```python
#Show only the columns I need and remove asterisks from team name
nfl_data = nfl_data[["Year", "League", "Team", "Wins", "Loses", "Ties", "Winning %"]]
nfl_data["Team"] = nfl_data["Team"].str.replace('*', '')
nfl_data.head()
```

```python
#Remove all seasons but 2000 - 2017
nfl_2000 = nfl_data.loc[nfl_data["Year"] > 1999]
nfl_seasons_2000_to_2017 = nfl_2000.loc[nfl_2000["Year"] < 2018]
nfl_seasons_2000_to_2017.head()
```

```python
pivot = nfl_seasons_2000_to_2017.pivot(values = "Winning %", index = "Team", columns = "Year")
pivot.head()
```
